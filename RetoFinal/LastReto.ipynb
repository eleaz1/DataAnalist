```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
! pip install imbalanced-learn > null
from imblearn.under_sampling import RandomUnderSampler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
URL = "https://raw.githubusercontent.com/it-ces/Datasets/main/basesuper.csv"
df = pd.read_csv(URL)
```

```python
df['ratio endeudamiento'] = (df['Pasivos corrientes totales']/df['Total pasivos'])*100
df['Razón de deuda'] = df['Total pasivos']/df['Total de activos']
df['Ratio de liquidez']= df['Activos corrientes totales']/df['Pasivos corrientes totales']
df['Prueba Acida'] = (df['Activos corrientes totales']- (df['Inventarios corrientes']+df['Inventarios no corrientes']))/df['Pasivos corrientes totales']
df['Capital de trabajo'] = df['Activos corrientes totales']- df['Pasivos corrientes totales']
df['Apalancamiento financiero'] = df['Total de activos']/df['Patrimonio total']
```

```python
X = ['Total pasivos', 'Ganancia (pérdida)','ratio endeudamiento','Ganancias acumuladas','Razón de deuda','Activos corrientes totales','Pasivos corrientes totales','Capital de trabajo']
df = df[X + ['y']] # una subase con las variables que me interesan X y el target
df.dropna(inplace=True)
df.isnull().sum()
```

```python
X = df[X]
y = df['y']
rus = RandomUnderSampler(random_state=23)
X_res, y_res = rus.fit_resample(X, y) # a better practice is pass X_train, y_train
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, shuffle = True, random_state = 666, stratify=y_res)
```

Esta sería la mejor opción de todas.

```python
clf  = DecisionTreeClassifier()
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
print(classification_report(y_test , preds))
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
classifier_lg = LogisticRegression()
classifier_lg.fit(X_train, y_train)
preds= classifier_lg.predict(X_test)
print(classification_report(y_test, preds))
```

```python
clasificador = RandomForestClassifier()
clasificador.fit(X_train, y_train)
preds = clasificador.predict(X_test)
print(classification_report(y_test,preds ))
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, confusion_matrix
from sklearn.metrics import f1_score  #ytrue, ypred
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score


def grid_lr(X_train, y_train):
    model = LogisticRegression(random_state=666, max_iter=1000)
    solvers = ['liblinear']
    penalty = ['l2','l1']
    c_values = [ 10, 1.0, 0.1, 0.01, 0.001, ]
    grid = dict(solver=solvers,penalty=penalty,C=c_values)
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=1, random_state=1)
    grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv,
                           scoring='roc_auc',error_score='raise')
    grid_result = grid_search.fit(X_train, y_train)
    return  grid_result.best_estimator_

classifier  = grid_lr(X_train, y_train)
classifier.fit(X_train, y_train)
preds = classifier.predict(X_test)
print(classification_report(y_test,preds ))
```
